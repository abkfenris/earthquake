```javascript
function fedcatalogToObject(input_string) {
    var lines = input_string.split('\n')
    var last_net = []
    var stations = {}
    for (var num in lines) {
        var line = lines[num]
        if (line === '') {
            last_net = []
        } else if (line.split(' ').length < 2) {
            last_net.push(line)
        } else {
            var name = line.split(/( (\d\d) | (--) )/)[0]
            stations[name] = {net: last_net}
        }
    }
    return (stations)
}
```
